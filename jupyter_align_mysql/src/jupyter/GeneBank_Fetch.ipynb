# Fetching

```python

GI = ['AF271693.1']
gb = Entrez.efetch(db="nuccore",id=GI,rettype="gb",retmode="text")
gb_file = gb.read()
gb.close()
print(gb_file)
```

```python
genome_id = '282598606'

record = Entrez.efetch(db="nuccore", id=genome_id, rettype="gb", retmode="text")

filename = 'out/genBankRecord.gb'
with open(filename, 'w') as f:
    f.write(record.read())
parsed_gb_file = next(SeqIO.parse(filename, "genbank"))

#print(parsed_gb_file)
print( parsed_gb_file.annotations["source"]) 
```

# Output in GB format

```python
GI = ['282598706']

for genome_id in GI:
    record = Entrez.efetch(db="nuccore", id=genome_id, rettype="gb", retmode="text")

    filename = 'generated/genBankRecord_{}.gb'.format(genome_id)
    print('Writing:{}'.format(filename))
    with open(filename, 'w') as f:
        f.write(record.read())
```
